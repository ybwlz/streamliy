为什么成交额和委托数量一正一负？

这是为了**会计记账和持仓统计**的需要，遵循的是**“方向”**逻辑，而不是“金钱流向”逻辑。

请看规律：

*   **开多 (Open Long)**：动作是**“买入”**。
    *   委托数量：`1` (增加持仓)
    *   成交额：`480,520` (正数，代表建立了正向的资产价值)
*   **平多 (Close Long)**：动作是**“卖出”**。
    *   委托数量：`-1` (减少持仓)
    *   成交额：`-481,320` (负数，代表消除了正向的资产价值)

**背后的逻辑：**

1.  **数量对冲 (Netting)**：
    系统通过加总数量来计算你现在手里还有多少货。
    $$ 1 (\text{开}) + (-1) (\text{平}) = 0 (\text{当前空仓}) $$
    如果都是正数，系统会以为你买了 2 手。

2.  **金额跟随数量**：
    在大多数交易系统的原始日志中：
    $$ \text{成交额} = \text{成交价} \times \text{成交数量} $$
    *   开仓时：$480.52 \times 1 = 480,520$
    *   平仓时：$481.32 \times (-1) = -481,320$

**特别注意：这里的负数成交额不代表你亏钱！**

你看第二行（2024/1/11）：
*   成交额是 **-481,320** (负很大)
*   平仓盈亏是 **800** (正赚钱)

**它是怎么算出来的？**
$$ \text{盈亏} = (\text{卖出价} - \text{买入价}) \times \text{数量} \times \text{合约乘数} $$
或者从流水看：
$$ \text{盈亏} \approx -(\text{开仓成交额} + \text{平仓成交额}) $$
*(注：这里符号取决于具体系统的定义，有些系统卖出是正现金流。但在你的这份数据里，你只需要关注 **“平仓盈亏”** 这一列即可，成交额的正负只是代表交易方向。)*

### 总结

1.  **关于 $P_{start}$**：不要用第一笔的成交额。请手动设定一个合理的初始资金（例如 20万 或 50万），足以覆盖保证金即可。
2.  **关于正负号**：
    *   **正数** = 买入/建仓
    *   **负数** = 卖出/平仓
    *   计算收益时，**直接累加“平仓盈亏”列**，不要去管“成交额”列的正负。
