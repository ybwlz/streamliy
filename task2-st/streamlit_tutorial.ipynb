````python
# Streamlit 是 Python 生态中专注于「快速数据可视化与 Web 应用开发」的框架，
# 核心特点：零前端代码（无需 HTML/CSS/JS）、语法极简、实时热重载，
# 专为数据科学家/分析师设计，支持数据展示、交互组件、图表绘制、文件处理等，
# 10行代码即可搭建数据应用，适合快速原型开发、数据看板、模型演示、数据分析报告
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------- 1. 基础页面元素（文本/标题）----------------------
# 页面标题（支持Markdown语法）
st.title("Streamlit 基础使用笔记")  # 一级标题
st.header("1. 基础页面元素")       # 二级标题
st.subheader("1.1 文本展示")       # 三级标题
st.markdown("**加粗文本**、*斜体文本*、[链接](https://streamlit.io/)")  # Markdown
st.text("普通纯文本内容，不解析Markdown")  # 纯文本
st.success("成功提示信息")  # 成功状态提示
st.warning("警告提示信息")  # 警告状态提示
st.error("错误提示信息")    # 错误状态提示
st.info("普通信息提示")     # 信息提示

# ---------------------- 2. 数据展示（核心功能）----------------------
st.header("2. 数据展示")
# 示例数据（DataFrame）
df = pd.DataFrame({
    "日期": pd.date_range("2025-01-01", periods=5),
    "收盘价": np.random.randint(7000, 7500, 5),
    "RSI": np.random.randint(20, 80, 5),
    "买入点": [True, False, True, False, False],
    "卖出点": [False, True, False, True, False]
})

# 展示DataFrame（支持排序、筛选、分页）
st.dataframe(df, use_container_width=True)  # use_container_width：自适应宽度
# 展示静态表格（无交互功能，纯展示）
st.table(df.head(3))  # 只展示前3行
# 展示JSON格式数据
st.json(df.iloc[0].to_dict())  # 展示第一行数据的JSON格式

# ---------------------- 3. 交互组件（用户输入）----------------------
st.header("3. 交互组件")
# 3.1 按钮（触发简单逻辑）
if st.button("点击获取随机数"):
    st.write(f"随机数：{np.random.randint(1, 100)}")

# 3.2 文本输入框
name = st.text_input("请输入姓名", value="数据分析师")  # value：默认值
st.write(f"欢迎你，{name}！")

# 3.3 滑块（数值选择）
period = st.slider("选择RSI周期", min_value=5, max_value=30, value=14, step=1)
st.write(f"当前RSI周期：{period}")

# 3.4 日期选择器（单个日期/日期范围）
single_date = st.date_input("选择单个日期")
date_range = st.date_input("选择日期范围", value=[pd.Timestamp("2025-01-01"), pd.Timestamp("2025-01-10")])
st.write(f"单个日期：{single_date}，日期范围：{date_range}")

# 3.5 下拉选择框
option = st.selectbox("选择数据类型", options=["收盘价", "RSI", "成交量"])
st.write(f"当前选择：{option}")

# 3.6 复选框（控制内容显示/隐藏）
if st.checkbox("显示原始数据"):
    st.dataframe(df)

# 3.7 单选按钮
radio_option = st.radio("选择展示图表类型", options=["折线图", "柱状图", "散点图"])

# ---------------------- 4. 数据可视化（图表绘制）----------------------
st.header("4. 数据可视化")
# 4.1 Streamlit原生图表（极简语法，无需额外配置）
st.subheader("4.1 原生图表")
# 折线图
st.line_chart(df, x="日期", y="收盘价", use_container_width=True, color="#1f77b4")
# 柱状图
st.bar_chart(df, x="日期", y="RSI", use_container_width=True, color="#ff7f0e")

# 4.2 结合Matplotlib绘图（支持自定义样式）
st.subheader("4.2 Matplotlib集成")
plt.figure(figsize=(10, 4))
# 绘制收盘价折线+RSI柱状图
ax1 = plt.subplot(111)
ax1.plot(df["日期"], df["收盘价"], label="收盘价", color="blue")
ax1.set_ylabel("收盘价", color="blue")
ax2 = ax1.twinx()  # 双Y轴
ax2.bar(df["日期"], df["RSI"], alpha=0.5, label="RSI", color="orange")
ax2.set_ylabel("RSI", color="orange")
plt.title("收盘价与RSI趋势图")
st.pyplot(plt)  # 直接渲染Matplotlib图表

# ---------------------- 5. 文件上传与处理 ----------------------
st.header("5. 文件上传与处理")
# 上传CSV/Excel文件
uploaded_file = st.file_uploader("上传数据文件（CSV/Excel）", type=["csv", "xlsx"])
if uploaded_file is not None:
    # 读取CSV文件
    if uploaded_file.name.endswith(".csv"):
        upload_df = pd.read_csv(uploaded_file)
    # 读取Excel文件
    else:
        upload_df = pd.read_excel(uploaded_file)
    # 展示上传的数据
    st.dataframe(upload_df, use_container_width=True)
    # 快速统计分析
    st.subheader("数据统计信息")
    st.write(upload_df.describe())

# ---------------------- 6. 页面布局（分栏/容器）----------------------
st.header("6. 页面布局")
# 6.1 左右分栏（columns）
col1, col2 = st.columns(2)  # 2列布局（可设比例：st.columns([1,2])）
with col1:
    st.subheader("左栏：买入点统计")
    buy_count = df["买入点"].sum()
    st.metric("买入点数量", buy_count)  # 指标卡片
with col2:
    st.subheader("右栏：卖出点统计")
    sell_count = df["卖出点"].sum()
    st.metric("卖出点数量", sell_count)

# 6.2 容器（Container）- 分组管理元素
with st.container():
    st.subheader("容器内内容（分组展示）")
    st.write("可以将相关元素放在一个容器中，便于管理")
    st.dataframe(df[["日期", "买入点", "卖出点"]])

# ---------------------- 7. 核心运行方式 ----------------------
# 1. 保存文件为 streamlit_basic.py
# 2. 终端执行命令：streamlit run streamlit_basic.py
# 3. 自动打开浏览器，实时预览（修改代码保存后自动刷新）

# ---------------------- 8. 核心优势总结 ----------------------
# - 零前端成本：纯Python代码，无需HTML/CSS/JS
# - 实时热重载：修改代码保存后，页面自动刷新
# - 极简语法：一行代码实现图表、交互组件
# - 数据友好：原生支持Pandas、NumPy、Matplotlib等数据科学库
# - 快速部署：支持Streamlit Cloud、Heroku等平台一键部署
````

### 1. 关于 `__enter__` 和 `__exit__`
你说的非常对。在 Python 中，任何能跟在 `with` 后面的对象，其类内部必然实现了两个魔术方法：
*   **`__enter__()`**：进入 `with` 块时执行。
*   **`__exit__()`**：退出 `with` 块时执行。

**在 Streamlit 的源码逻辑里，大概发生了这样的事情（简化版）：**

1.  **遇到 `with st.sidebar:`**：
    *   触发 `__enter__`。
    *   系统把“当前正在写入的容器”由 **主页面** 切换为 **侧边栏**。
    *   同时，系统内部会把之前的环境（主页面）**压入一个栈（Stack）里“暂存”起来**。

2.  **执行 `with` 内部代码**：
    *   所有的 `st.header`、`st.write` 都会写入当前顶层的容器（也就是侧边栏）。

3.  **退出 `with` 块**：
    *   触发 `__exit__`。
    *   系统把侧边栏从栈顶移走。
    *   系统**弹出**之前“暂存”的主页面环境，恢复它作为当前容器。

### 2. 关于生成器（Generator）的比喻
你的直觉很敏锐。虽然技术实现上：
*   **生成器** (`yield`) 主要是为了保存函数的**执行进度和变量状态**。
*   **上下文管理器** (`with`) 主要是为了管理**外部资源或环境上下文**（比如文件开关、数据库连接、或者这里的UI布局区域）。

但从**“流程控制”**的角度看，它们确实非常像：**都是先挂起当前状态，去处理别的事情，处理完再无缝接轨回来。**

### 总结
你现在的注释逻辑可以这样理解：

```python
with st.sidebar: 
    # 1. __enter__: 暂存主页面光标，将光标“瞬移”进侧边栏容器
    st.header("...")
    # ...
    # 2. __exit__:  with结束，自动将光标“恢复”回主页面位置
```

所以，你现在的理解比最开始的更准确，完全可以自信地认为它是**“保存主页环境 -> 切换进侧边栏 -> 恢复主页环境”**的过程。
